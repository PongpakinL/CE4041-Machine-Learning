```python
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv('train_modelling_rfe.csv')
train
```

```python
test = pd.read_csv('test_modelling_rfe.csv')
test
```

```python
df = pd.read_csv('submission.csv')
df
```

```python
def cluster_and_predict(k):
    X = train.drop("price_doc", axis=1)
    y = train["price_doc"]

    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    X['cluster_label'] = kmeans.labels_

    linear_regression_models = {}

    for cluster_label in range(k):
        X_cluster = X[X['cluster_label'] == cluster_label].drop('cluster_label', axis=1)
        y_cluster = y[X['cluster_label'] == cluster_label]
        reg = LinearRegression()
        reg.fit(X_cluster, y_cluster)
        linear_regression_models[cluster_label] = reg

    test['predicted_cluster_label'] = kmeans.predict(test)
    y_pred = test['predicted_cluster_label'].map(lambda x: linear_regression_models[x].predict(test[test['predicted_cluster_label'] == x].drop(['predicted_cluster_label'], axis=1))[0])
    
    return y_pred
```

```python
y_pred = cluster_and_predict(1)
```

```python
df['price_doc'] = y_pred
df
```

```python
df.to_csv("Kmean_Linear_New.csv", index=False)
```
